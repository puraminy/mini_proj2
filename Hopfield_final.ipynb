##Weight

```
def weight(X):
  X = np.asarray(X)
  n,m = X.shape
  W=np.zeros([m,m])
  for i in range(n):
    W =W + np.matmul(np.transpose(np.asmatrix(X[i,:])),np.asmatrix(X[i,:]))
    W = W-np.diag(np.diag(W))
  
  return W

W = weight(X)

import matplotlib.pyplot as plt

import seaborn as sn
import pandas  as pd

df_W = pd.DataFrame(W, range(64),range(64))
#axlabels = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']

plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_W, annot=True,annot_kws={"size": 12}, xticklabels=[], yticklabels=[])# font size
plt.show()
np.min(W)
```

##Hopfield

```
def Hop(IN,W):
  x = np.expand_dims(IN, axis = 0)
  y = x
  n=len(IN)
  y2=y-1
  while np.max(np.abs(y2-y))!=0 or np.min(np.abs(y2-y))!=0:
    y2=y
    for i in range(n):
      y_ini = x[0,i] + y*W[:,i]
      if y_ini[0,0]>0:
        y[0,i] = 1
      elif y_ini[0,0]<0:
        y[0,i] = -1
      
  return y
fig, axes = plt.subplots(2, 10)
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
plt.axis('off')

for i in range(len(X)):
  IN=X[i]
  y=Hop(IN,W)

  axes[0,i].imshow(-1*np.reshape(IN,[8,8]))
  axes[0,i].axis('off')
  axes[1,i].imshow(-1*np.reshape(y,[8,8]))
  axes[1,i].axis('off')
```

##Noise

```
import random
import scipy.spatial.distance as dis
def test_noise (inp, te_n, n_n): #inp:input vector te_n: number of noisy samples for test. n_n: number of noisy elements
  test = []
  for i in range(te_n):
    rnd = random.sample(range(len(inp)), n_n)
    noise = np.asarray(inp)
    for i in rnd:
      noise[i] = -1*noise[i]
    test.append(noise)
    
  test=np.asarray(test)
  test = np.unique(test,axis=0)
  test=np.ravel(test)
  return test

def accuracy(test, t, W): #test matrix, t: output, W: Weight
  j=0
  n=len(test[:,0])
  for i in range(n):
    y=Hop(test[i,:],W)
    if dis.hamming(y, t)==0:
      j=j+1
  acc=j/n
  return acc

plt.figure
fig, axes = plt.subplots(3, len(X))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
plt.axis('off')

for i in range(len(X)):
  IN=test_noise (X[i], 1, 20)
  y=Hop(IN,W)

  axes[0,i].imshow(-1*np.reshape(X[i],[8,8]))
  axes[0,i].axis('off')
  axes[1,i].imshow(-1*np.reshape(IN,[8,8]))
  axes[1,i].axis('off')
  axes[2,i].imshow(-1*np.reshape(y,[8,8]))
  axes[2,i].axis('off')
  
```

```
XX = [X[0],X[1], X[2]]
W1 = weight(XX)

fig, axes = plt.subplots(2,len(XX))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)

for i in range(len(XX)):
  axes[0,i].imshow(-1*np.reshape(XX[i],[8,8]))
  axes[0,i].axis('off')
  axes[1,i].imshow(-1*np.reshape(Hop(XX[i],W1),[8,8]))
  axes[1,i].axis('off')

acc=[]
for i in range(len(XX)):
  te=test_noise (XX[i], 1000, 20)
  acc.append(accuracy(te, X[i], W1))
print('acc =' ,acc)
```

```
import matplotlib.pyplot as plt
import scipy.spatial.distance as dis

hammX = np.zeros((len(X),len(X)))
for i in range(len(X)):
  for j in range(len(X)):
    hammX[i,j] = dis.hamming(X[i],X[j])
    
# Visualizing of Hamming Distance
import seaborn as sn
import pandas  as pd

df_hammX = pd.DataFrame(hammX, range(len(X)),
                  range(len(X)))
axlabels = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']

plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_hammX, annot=True,annot_kws={"size": 12}, xticklabels=axlabels, yticklabels=axlabels)# font size
plt.show()
```
